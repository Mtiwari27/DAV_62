Aim : Explore Data Visualization libraries in Python and R

Installed libraries : ggplot2, lattice and RColorBrewer

```r
install.packages("ggplot2")
library(ggplot2)
```

```r
install.packages("RColorBrewer")
library(RColorBrewer)
```

```r
install.packages("lattice")
library(lattice)
```

Loaded the Iris Dataset

```r
data(iris)
head(iris)
```

Custom colour palette using RColorBrewer

```r
custom_palette <- brewer.pal(3, "Set1")
```

Data Visualization using ggplot2

```r
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) +
  geom_point() +
  scale_color_manual(values = custom_palette) +
  labs(x = "Sepal Length", y = "Sepal Width", title = "Scatter Plot of Sepal Length vs Sepal Width") +
  theme_minimal()
```

Data Visualization using Lattice

```r
xyplot(Sepal.Width ~ Sepal.Length, data = iris,
       xlab = "Sepal Length", ylab = "Sepal Width",
       main = "Scatter Plot using Lattice")
```
