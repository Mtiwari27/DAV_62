Implementation of simple linear regression in R

```r
#loading libraries
install.packages("ggplot2")
install.packages("dplyr")
install.packages("broom")
```

```r
library(ggplot2)
library(dplyr)
library(broom)
```

```r
data <- read.csv('/content/Salary_dataset.csv')
model <- lm(Salary ~ YearsExperience, data = data)
summary(model)
```

```r
#scatter plot with regression line
ggplot(data, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Scatter Plot with Regression Line",
       x = "YearsExperience",
       y = "Salary")
```

```r
install.packages("car")
library(car)
```

```r
install.packages("carData")
library(carData)
```

```r
#Normality Check
qqPlot(model$residuals, main="Q-Q Plot of Residuals")
```

```r
#Linearity Check
crPlots(model, main = "Residual Plot for Linearity Check")
```
