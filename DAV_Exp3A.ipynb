Packages/Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Model creation and model fitting

```python
X = df.drop('income', axis=1)
y = df['income']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

Coefficient of Determination, Intercept and Coefficients

```python
#Coefficient of Determination
r2_score = model.score(X_test, y_test)
print(f'Coeff. of Det is : {r2_score}')
#Intercept
intercept = model.intercept_
print(f'Intercept is : {intercept}')
#Coefficients
coefficient = model.coef_
print(f'Coefficients are:')
for feature, coef in zip(X.columns, coefficient):
  print(f'{feature} : {coef}')

#Equation of line
equation = f"y = {intercept:.2f} + "
for feature, coef in zip(X.columns, coefficient):
  equation += f"{coef:.2f} * {feature} + "

equation = equation[:-2]
print(f"Regression Equation: {equation}")
```
